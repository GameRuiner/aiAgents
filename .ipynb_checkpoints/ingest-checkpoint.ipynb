```python
from agentlib.knowledge_loader import KnowledgeLoader
from minsearch import Index, VectorSearch
from sentence_transformers import SentenceTransformer
```

```python
loader = KnowledgeLoader("google", "adk-python")
docs = loader.load_repository()
```

```python
f"ADK documents: {len(docs)}"
```

```python
docs[0]
```

```python
adk_chunks = []

for doc in docs:
    doc_copy = doc.copy()
    doc_content = doc_copy.pop('content')
    sections = loader.chunk_by_headers(doc_content, level=2)
    for section in sections:
        section_doc = doc_copy.copy()
        section_doc['section'] = section
        adk_chunks.append(section_doc)
```

```python
len(adk_chunks), adk_chunks[0]
```

```python
index = Index(
    text_fields=["name", "section", "filename"],
    keyword_fields=[]
)

index.fit(adk_chunks)
```

```python
embedding_model = SentenceTransformer('multi-qa-distilbert-cos-v1')
```

```python
from tqdm.auto import tqdm
import numpy as np

adk_embeddings = []

for d in tqdm(adk_chunks):
    text = d['filename'] + ' ' + d['section']
    v = embedding_model.encode(text)
    adk_embeddings.append(v)

adk_embeddings = np.array(adk_embeddings)
```

```python
vindex = VectorSearch()
vindex.fit(adk_embeddings, adk_chunks)
```

```python
query = 'I just found out about the course. Can I enroll now?'
v_query = embedding_model.encode(query)
vector_results =vindex.search(v_query, num_results=5)
```

```python
text_results = index.search(query, num_results=5)
```

```python
final_results = text_results + vector_results
```
